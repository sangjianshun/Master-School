```python
import pandas as pd
import numpy as np
```

```python
file_address = "data/chipotle.tsv"
chipo = pd.read_csv(file_address, sep='\t')
```

```python
chipo.head(30)
```

单价超过10的item一共有多少个。（只考虑购买数量是1的情况）

价格指的是单价

需要去重

```python
prices = [float(value[1 : -1]) for value in chipo.item_price]
chipo.item_price = prices
chipo_one_prod = chipo[chipo.quantity == 1]
chipo_one_prod.query('item_price > 10').item_name.nunique()
```

每一个产品的价格是多少？

```python
chipo_one_prod = chipo_one_prod.drop_duplicates(["item_name", "item_price"]) #按照item_name和item_price进行去重
chipo_one_prod[["item_name", "item_price"]]
```

按照item_name进行排序，看一个item_name是否有多个价格

```python
print(chipo.item_name.sort_values())
print("##########")
chipo_one_prod.sort_values(by = "item_name")
```

最贵的单价是什么商品

```python
chipo.sort_values(by = "item_price", ascending = False).head(1)
```

商品Veggie Salad Bowl出现在多少订单里

```python
chipo_salad = chipo[chipo.item_name == "Veggie Salad Bowl"]
print(len(chipo_salad))
```

统计商品“Canned Soda”被订购超过一份的订单有多少个

```python
chipo_drink_steak_bowl = chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)]
print(len(chipo_drink_steak_bowl))
print("#############")
chipo_drink_steak_bowl.head()
```

# 第二个数据集

```python
file_address = "data/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(file_address, sep=',')
euro12.head()
```

一共有多少支队伍参加比赛

```python
print(euro12.Team.nunique())
print("##########")
print(euro12.shape[0]) # 说明什么？
```

将球队名称设置为index

```python
euro12.set_index("Team",inplace=True)
euro12.head()
```

只看一支队伍的分数

```python
print(euro12[['Goals']].head()) #注意和一个中括号的区别
print("##############")
print(euro12['Goals'].head())
print("##############")
print(euro12.Goals.head())
print("##############")
print(type(euro12[['Goals']]), type(euro12['Goals']))
```

只看一支队伍对应的黄牌数，红牌数

```python
discipline = euro12[['Yellow Cards', 'Red Cards']]
discipline
```

查看队伍犯规比较厉害的是哪些？（按照红牌，黄牌进行排序）

```python
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending = False)
```

查看本次比赛裁判对规则判罚的严厉程度（可以用平均每只队伍黄牌的获取张数）

```python
print(discipline['Yellow Cards'].mean())
print("###########")
print(discipline.describe())
print("###########")
discipline.describe()["Yellow Cards"]["mean"] # 为什么反过来["Yellow Cards"]["mean"]会报错，这涉及到dataframe的索引方式。
```

过滤掉得分小于等于6个的球队（只取得分超过6个的球队）

```python
euro12[euro12.Goals > 6]
```

取球队名称以“G”开头的球队（字符串以某字符开头的逻辑）

```python
euro12[euro12.index.str.startswith('G')] # 注意此时列Team已经是dataframe的index了，不能使用euro12.Team进行索引
```

选择表的前面四列（注意head是选择行）

```python
euro12.iloc[: , 0:4]
```

```python
euro12.head()
```

只看来自于球队England，Italy，Russia的指标Shooting Accuracy（注意区分loc和iloc）

```python
print(euro12.loc[euro12.index.isin(['England', 'Italy', 'Russia']), ['Shooting Accuracy']])
print("######")
print(euro12.loc[['England', 'Italy', 'Russia'], ['Shooting Accuracy']])
```
