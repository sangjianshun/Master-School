## loc和iloc的索引形式是一致的，区别在于loc是从行或列的名称出发，iloc是从行或列的位置出发。
### 1. 只索引其中的行
返回是series
### 2. 索引行和列
基础格式是[行,列]，返回dataframe
### 3. 传入布尔类型
布尔类型对应的list需要和dataframe的长度保持一致

```python
import pandas as pd
```

```python
file_address = "data/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(file_address, sep=',')
euro12.set_index("Team", inplace = True)
euro12.head(10)
```

# 1. loc
### 1.1. loc只索引其中的行
loc直接填写索引的名称，即选择特定一行，返回series数据类型

留一个思考：如果想让某一列变成series数据类型怎么办（我们在前面的视频讲到过）

```python
print(type(euro12.loc["England"]))
euro12.loc["England"]
```

### 1.2. loc索引行和列
基础格式是loc[行,列]，返回dataframe

行和列可以是list比如["x1"]或者["x1","x2"]，可以是切片比如"x1":"x2"或者:   

(注意这里使用切片的方式就不需要外面再加一层中括号了)

```python
euro12.loc[["England"],["Goals","Shooting Accuracy"]]
```

```python
euro12.loc[["England", "Italy"], "Goals":"Shooting Accuracy"]
```

### 下面这三种情况是等价的
euro12.loc[["England"]]

euro12.loc[["England"],:]

euro12.loc[["England"],]

```python
euro12.loc[["England"]]
```

```python
euro12.loc[["England"],:]
```

```python
euro12.loc[["England"],]
```

### 如果去索引列，前面无法省略

```python
euro12.loc[:,["Goals"]]
```

### 1.3. loc可以传入布尔类型
布尔类型的长度需要和原数据和行数相等

```python
print(~euro12.index.isin(['England', 'Italy']))
print("##########")
euro12.loc[~euro12.index.isin(['England', 'Italy']), :]
```

```python
euro12.loc[['England', 'Italy'], euro12.columns.isin(["Goals", "Yellow Cards"])]
```

# 2. iloc
### 2.1. iloc只索引其中的行
iloc直接填写索引的位置，即选择特定一行，返回series数据类型

```python
euro12.head(10)
```

```python
euro12.iloc[3]
```

### 2.2. iloc索引行和列
基础格式是iloc[行,列]，返回dataframe

行和列可以是list比如[3]或者[3,5]，可以是切片比如27:32或者:   

(注意这里使用切片的方式就不需要外面再加一层中括号了)

```python
euro12.iloc[[3,5], [27,29,30]]
```

```python
euro12.iloc[[3,5], 27:32]
```

### 下面这三种情况是等价的
euro12.iloc[[3]]

euro12.iloc[[3],:]

euro12.iloc[[3],]

```python
euro12.iloc[[3]]
```

```python
euro12.iloc[[3],:]
```

```python
euro12.iloc[[3],]
```

### 2.3. iloc可以传入布尔类型
布尔类型的长度需要和原数据和行数相等

```python
print(~euro12.index.isin(['England', 'Italy']))
print("##########")
euro12.iloc[~euro12.index.isin(['England', 'Italy']), :]
```
